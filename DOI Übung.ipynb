```python
base_url = "https://api.crossref.org/works/"

for doi in dois:
    print(doi)
    full_url = base_url + doi
    print(full_url)
    doi_json_data = urllib.request.urlopen(full_url).read()
    doi_data = json.loads(doi_json_data)
    article_titel = doi_data["message"]["title"][0]
    journal_titel = doi_data["message"]["short-container-title"][0]
    print(article_titel)
    print(journal_titel)                       
```
